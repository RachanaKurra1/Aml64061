```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np

centers = [[2, 4], [6, 6], [1, 9]]
n_classes = len(centers)
data, labels = make_blobs(n_samples=600, 
                          centers=np.array(centers),
                          random_state=1)
print(data, labels)
```

```python
# import modules for this project
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#training testing split
split_data = train_test_split(data, labels, 
                       train_size=0.8,
                       test_size=0.2,
                       random_state=12)
train_data, test_data, train_labels, test_labels = split_data
print(split_data)
```

```python
# Create and fit a nearest-neighbor classifier
from sklearn.neighbors import KNeighborsClassifier
# classifier "out of the box", no parameters
knn = KNeighborsClassifier()
knn.fit(train_data, train_labels) 
```

```python
# print some interested metrics
print("Predictions from the classifier:")
learn_data_predicted = knn.predict(train_data)
print(learn_data_predicted)
print("Target values:")
print(train_labels)
print(accuracy_score(learn_data_predicted, train_labels))
```

```python
# re-do KNN using some specific parameters. 
knn2 = KNeighborsClassifier(algorithm='auto', 
                            leaf_size=30, 
                            metric='minkowski',
                            p=2,         # p=2 is equivalent to euclidian distance
                            metric_params=None, 
                            n_jobs=1, 
                            n_neighbors=5, 
                            weights='uniform')

knn.fit(train_data, train_labels) 
test_data_predicted = knn.predict(test_data)
accuracy_score(test_data_predicted, test_labels)

```

Plotting the **results** on accuracy

```python
number_of_neighbors = np.arange(1, 10)
train_accuracy = np.empty(len(number_of_neighbors))
test_accuracy = np.empty(len(number_of_neighbors))

for i, k in enumerate(number_of_neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    # Fit the classifier to the training data
    knn.fit(train_data,train_labels)
    
    # we calculate  accuracy on the training set and testing set
    train_accuracy[i] = knn.score(train_data,train_labels)
    test_accuracy[i] = knn.score(test_data,test_labels)

# plotting graph on accuracy and KNN

plt.title('k-NN:  Number of Neighbors')
plt.plot(number_of_neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(number_of_neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```
